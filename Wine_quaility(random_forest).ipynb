```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv('/content/drive/MyDrive/winequality-red.csv')
```

```python
x = data.drop(columns = 'quality',axis = 1)
```

```python
y = data['quality']
```

```python
print(x)
```

```python
print(y)
```

```python
sns.countplot(x='quality', data=data)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
# Instantiate the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)

# Train the model
rf_classifier.fit(x_train, y_train)

# Make predictions
predictions = rf_classifier.predict(x_test)

# Evaluate the model
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)
#Remember that while Random Forests are powerful, they might not always be the best choice for every dataset or problem. It's a good practice to experiment with different algorithms to find the one that works best for your specific task.

```
